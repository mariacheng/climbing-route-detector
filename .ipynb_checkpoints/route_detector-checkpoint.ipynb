## Detect Climbing Routes from an Image



```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('testimage.jpg')
plt.imshow(img)
```

```python
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
hist = cv.calcHist([img_hsv], [0], None, [180], [0,180])
plt.plot(hist)
```

```python
# Detect red
mask1 = cv.inRange(img_hsv,(0, 100, 20), (5, 255, 255) )
mask2 = cv.inRange(img_hsv,(175,50,20), (180,255,255) )
mask = cv.bitwise_or(mask1, mask2)

img_red = cv.bitwise_and(img, img, mask=mask)
plt.imshow(img_red)
```
